```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
students = np.array([
    "ibrahim","Abubakar","Charan","Farhan","Kasim",
    "Jabeer","Nawaz","Arjun","","Kumar",
    "Ajay","Vikram","Teja","Ram","Ali"
])
```

```python
subjects = np.array(["Python", "Java", "DBMS", "AI", "Agentic_Ai"])
```

```python
marks = np.array([
    [72,47,44,51,40],
    [50,46,47,80,41],
    [47,42,59,55,55],
    [46,60,72,53,44],
    [58,84,63,44,98],
    [92,48,64,42,58],
    [36,70,48,92,44],
    [52,99,48,36,95],
    [76,42,85,49,42],
    [87,41,57,36,43],
    [57,83,78,61,64],
    [96,48,91,44,68],
    [84,48,97,39,86],
    [54,42,58,55,77],
    [80,77,39,73,58]
])
```

```python
#subject wise avg
subject_avg = np.mean(marks, axis=0)
print(subject_avg)
```

```python
# Student Wise Total
total = np.sum(marks, axis=1)
```

```python
# LINE PLOT
plt.figure(figsize=(8,4))
plt.plot(subjects, subject_avg, marker='o')
plt.title("Subject Wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.grid(True)
plt.show()
```

```python
# BAR GRAPH ----------------
plt.figure(figsize=(8,4))
plt.bar(subjects, subject_avg)
plt.title("Average Marks Bar Graph")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

```python
# HSTOGRAM ----------------
plt.figure(figsize=(8,4))
plt.hist(total, bins=5)
plt.title("Total Marks Distribution")
plt.xlabel("Total Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
#SCATTER PLOT ----------------
python_marks = marks[:,0]
math_marks = marks[:,4]

plt.figure(figsize=(8,4))
plt.scatter(python_marks, math_marks)
plt.title("Python vs Math Marks")
plt.xlabel("Python Marks")
plt.ylabel("Math Marks")
plt.grid(True)
plt.show()
```
