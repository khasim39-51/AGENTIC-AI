```python
#“Student Performance Analysis Using Data Visualization”
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sname=['kim','john','van','jay','hai']
sscore=[45,68,87,76,84]
sid=[1,2,3,4,5]
plt.figure(figsize=(20,2))
```

```python
plt.plot(sname,sscore,marker="s",label="sname",color="blue")
plt.plot(sname, sscore, marker="^", label="sscore" ,color="red")
#plt.legend();
plt.xlabel('sname')
plt.ylabel('sscore')
plt.title('Student Performance Analysis')
plt.bar(sname,sscore)
plt.grid(True,alpha=0.5)
plt.savefig("Student Performance Analysis.png")
plt.show()
```
