```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
age=[1,2,3,4,5,6,7]
salary=[100,200,300,400,500,600,700]
id=[1000,2000,3000,4000,5000,6000,7000]
plt.figure(figsize=(3,3))
plt.plot(age,salary,marker='s',label='age',color='blue')
plt.plot(age,id,marker='^',label='id',color='red')
plt.legend();
plt.xlabel('salary')
plt.ylabel('age')
plt.title('age vs salary')
plt.grid(True,alpha=0.3)
plt.savefig('plot.png')
plt.show()

```
